```python
from __future__ import unicode_literals
from polyglot.text import Text
import collections
import pandas as pd
import numpy as np
import re
import pickle
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split, KFold
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
from reldi_tokeniser import tokeniser
from reldi_tagger import tagger


# -*- coding: utf-8 -*-

from IPython.core.display import display, HTML
display(HTML("<style>.container { width: 95% !important; }</style>"))
```

```python
data = pd.read_csv('SerbMR-3C.csv', encoding='utf8')
```

```python
stupid_python_and_encodings_and_tools = {
    "\u0161" : "s",
    "\u017e" : "z",
    "\u010d" : "c",
    "\u0107" : "c"    
}

def removeStupidity(text):
    ret = ""
    for char in text.lower():
        if char in stupid_python_and_encodings_and_tools:
            ret += stupid_python_and_encodings_and_tools[char]
        else: ret += char
    return ret

def removePunctuation(asd):
    reg = re.compile(r'[^\w\s]')
    new = []
    for x in asd:
        if not reg.search(x):
            new.append(x.lower())
    return new
# for row in data.iterrows():
# text = removeStupidity(data['Text'][1543])
tokens_out = []
for row in data.iterrows():
# print(text)
    text, polarity = row[1]['Text'], row[1]['class-att']
    asd = tokeniser.tok(text, 'sr')
    asd = removePunctuation(asd.split('\n')[:-1])
#     print(asd)
    if polarity == 'POSITIVE':
        polarity = 1
    elif polarity == 'NEGATIVE':
        polarity = -1
    else: polarity = 0
    tokens_out.append((asd, polarity))

pickle.dump(tokens_out, open('tokenised_SerbMR-3C.pickle', 'wb'), 2)

# lam = []
# for word in asd:
#     lam.append(tagger.lemantise([word], 'sr')[0])
# print(lam)
```

```python
examples = []
labels = []
kfold = KFold(10, True, 44123)

for i, row in data.iterrows():
    examples.append(row['Text'])
    labels.append(row['class-att'])
    
labels = [1 if x == 'POSITIVE' else 0 for x in labels]
vectorizer = CountVectorizer(analyzer='word', lowercase=False)
features = vectorizer.fit_transform(examples)
features_nd = features.toarray()

log_model = MultinomialNB()

#acc = []
#for tr_idx, tst_idx in kfold.split(features_nd):
#    X_train, X_test, y_train, y_test = features_nd[tr_idx], features_nd[tst_idx], labels[tr_idx], labels[tst_idx]
#    log_model = log_model.fit(X=X_train, y=y_train)
#    y_pred = log_model.predict(X_test)
#    acc.append(accuracy_score(y_test, y_pred))

X_train, X_test, y_train, y_test = train_test_split(features_nd, labels, train_size=0.7, random_state=4459)
log_model = log_model.fit(X=X_train, y=y_train)
y_pred = log_model.predict(X_test)
acc.append(accuracy_score(y_test, y_pred))

print(np.mean(acc))
```
